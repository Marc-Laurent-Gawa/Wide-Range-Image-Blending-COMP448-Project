# View Results

```python
import os
import glob
import math
import matplotlib.pyplot as plt
from PIL import Image

try:
    batch_size = int(input("Enter batch size (default 5): ") or "5")
    if batch_size <= 0:
        raise ValueError
except ValueError:
    print("Invalid batch size.")
    exit()

root = os.getcwd()
image_folder = os.path.join(root, "results", "result1-2")

image_paths = sorted(glob.glob(os.path.join(image_folder, "*.png")))

if not image_paths:
    print("No images found.")
    exit()

num_images = len(image_paths)
print(f"Found {num_images} images.")
print("Starting batch processing...\n")

for i in range(0, num_images, batch_size):



    batch_end = min(i + batch_size, num_images)
    batch = image_paths[i:batch_end]

    print(f"Processing batch {i+1} to {batch_end}...")

    num_in_batch = len(batch)
    rows = math.ceil(math.sqrt(num_in_batch))
    cols = math.ceil(num_in_batch / rows)

    fig, axes = plt.subplots(rows, cols, figsize=(12, 8))
    axes = axes.flatten()

    for idx, img_path in enumerate(batch):
        print(f"  Loading: {os.path.basename(img_path)}")

        img = Image.open(img_path)

        axes[idx].imshow(img)
        axes[idx].set_title(os.path.basename(img_path))
        axes[idx].axis('off')

    for k in range(idx + 1, len(axes)):
        axes[k].axis('off')

    plt.suptitle(f"Batch {i+1} to {batch_end}", fontsize=16)
    plt.tight_layout()
    plt.show()

    print("Batch complete.\n")

    if i > 0:
      cont = input("Press ENTER for next batch or 'q' to quit: ")
      if cont.lower() == "q":
          print("Stopping early.")
          break

print("All batches finished.")
```
